```python
import csv
import 
import argparse
import requests
from datetime import datetime, timedelta
```

```python
date_range = (datetime.now().date() - timedelta(days=28), datetime.now().date())
number_of_calls = 10
movies = ["Megalopolis", "Wild Robot", "Joker 2", "Venom", "The Substance"]
output_file = "output.csv"
```

```python
URL = "https://api.newscatcherapi.com/v2/search"
API_KEY = "xk_ljaW7rQBcySn90-_yfy69PcS5QAkga-0cCjzfDMA"

headers = {
    "x-api-key": API_KEY
}

params = {
    "q": " OR ".join(movies),
    "from": date_range[0].strftime("%Y-%m-%d"),
    "to": date_range[1].strftime("%Y-%m-%d"),
    "lang": "en",
    "page_size": number_of_calls
}
    
response = requests.get(URL, headers=headers, params=params)
```

```python
if response.status_code == 200:
    data = response.json()
    with open(output_file, mode="w", newline="") as f:
        writer = csv.writer(f)
        writer.writerow(["Title", "Author", "Published Date", "Link", "Summary"])
            
        for article in data.get("articles", []):
            title = article.get("title")
            author = article.get("author")
            published_date = article.get("published_date")
            link = article.get("link")
            summary = article.get("summary")
                
            writer.writerow([title, author, published_date, link, summary])
else:
    print(f"Error: {response.status_code}.")
```
